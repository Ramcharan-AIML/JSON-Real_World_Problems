```python
import json
import numpy as np

api_response = '''
{
    "store": "TechMart",
    "location": "Hyderabad",
    "total_products": 3,
    "products": [
        {
            "id": 101,
            "name": "Laptop",
            "price": 75000,
            "in_stock": true,
            "specs": {
                "ram": "16GB",
                "storage": "512GB SSD"
            }
        },
        {
            "id": 102,
            "name": "Phone",
            "price": 25000,
            "in_stock": false,
            "specs": {
                "ram": "8GB",
                "storage": "128GB"
            }
        },
        {
            "id": 103,
            "name": "Tablet",
            "price": 40000,
            "in_stock": true,
            "specs": {
                "ram": "6GB",
                "storage": "256GB"
            }
        }
    ]
}
'''


# Convert this JSON string into a Python dictionary
data = json.loads(api_response)
print(data)

# Print the store name and location
store_loc = data["store"]
store_name = data["location"]
print(store_name,",", store_loc,"\n")

# Print the name and price of every product using a loop
for i  in data["products"]:
    print(i["name"],",", i["price"])

# Print only the products that are in stock
for prod in data["products"]:
    if prod["in_stock"]:
        print(prod)

# Print the RAM of the Laptop by accessing it directly (nested access)
lap_ram = data["products"][0]["specs"]["ram"]
print("The RAM of the Laptop:-",lap_ram)

# Find and print the most expensive product name
max = 0
prod_name = ""
for exp  in data["products"]:
    if exp["price"] > max:
        max = exp["price"]
        prod_name = exp["name"]
print(prod_name)

# Save this entire data to a file called store.json on your Desktop
with open ("data.json" , "w") as file:
    json.dump(api_response , file,  indent=4)

# Read it back from the file and print the total number of products
with open ("data.json" , "r") as file:
    loaded_data = json.load(file)

data_2 = json.loads(loaded_data)
print(len(data_2["products"]))
```

```python
import json
import pandas as pd
import numpy as np

api_response2 = '''
{
    "store": "TechMart",
    "location": "Hyderabad",
    "total_products": 5,
    "products": [
        {
            "id": 101,
            "name": "Laptop",
            "price": 75000,
            "in_stock": true,
            "rating": 4.5,
            "specs": {"ram": "16GB", "storage": "512GB SSD"}
        },
        {
            "id": 102,
            "name": "Phone",
            "price": 25000,
            "in_stock": false,
            "rating": 4.2,
            "specs": {"ram": "8GB", "storage": "128GB"}
        },
        {
            "id": 103,
            "name": "Tablet",
            "price": 40000,
            "in_stock": true,
            "rating": 3.9,
            "specs": {"ram": "6GB", "storage": "256GB"}
        },
        {
            "id": 104,
            "name": "Smartwatch",
            "price": 15000,
            "in_stock": true,
            "rating": 4.7,
            "specs": {"ram": "2GB", "storage": "32GB"}
        },
        {
            "id": 105,
            "name": "Headphones",
            "price": 8000,
            "in_stock": false,
            "rating": 4.1,
            "specs": {"ram": "N/A", "storage": "N/A"}
        }
    ]
}
'''

# Convert the JSON string to a dict and load only the products list into a pandas DataFrame
data = json.loads(api_response2)
print(data)

# Print the DataFrame and check its shape
df = pd.DataFrame.from_dict(data)
products_df = pd.json_normalize(data["products"])
print(df)
print(products_df)

# Find the average price of all products using numpy
# for price in data["products"]:
#     print(price["price"])
find_avg_price = np.mean([val["price"] for val in data["products"]])
print(find_avg_price)

# Find the highest rated product using pandas
find_rated_prod = products_df.loc[products_df["rating"].idxmax()]
# find_prod = products_df.iloc[3]
print(find_rated_prod)

# Filter and show only products that are in stock
filter_prod_stock = products_df.loc[products_df["in_stock"] == True]
print(filter_prod_stock)

# Add a new column called "discounted_price" which is 10% off the original price
products_df["discounted_price"] = (products_df["price"] / 10 )
print(data)

# Sort the DataFrame by price in descending order and print it
sort_df = products_df.sort_values(["price"] , ascending=False)
print(sort_df)

# Find the total value of all in-stock products combined (sum of their prices)
find_total_values = products_df.loc[products_df["in_stock"] == True , ["price"]].sum()
print(find_total_values)
```
